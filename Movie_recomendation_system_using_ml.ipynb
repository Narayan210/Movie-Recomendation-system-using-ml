```python
df=df.dropna()
```

```python
df.shape
```

```python
df['tags'] = df['genre']+ df['overview']
```

```python
df.head()
```

```python
new_df= df[['id', 'title', 'tags']]
```

```python
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000)
```

```python
cv
```

```python
print(new_df['tags'].isnull().sum())  # Check for nulls
```

```python
vec = cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vec)
```

```python
sim
```

```python
new_df[new_df['title']=='The Shawshank Redemption']
```

```python
dist = sorted(list(enumerate(sim[0])), reverse= True, key=lambda vec:vec[1])
```

```python
dist
```

```python
for i in dist [0:5]:
  print(new_df.iloc[i[0]].title)
```

```python
def recomend(df):
  index = new_df[new_df['title']==df].index[0]
  distance =sorted(list(enumerate(sim[index])), reverse= True, key=lambda vec:vec[1])
  for i in distance[0:5]:
    print(new_df.iloc[i[0]].title)
```

```python
recomend("Iron Man")
```
